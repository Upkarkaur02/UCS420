Q1. Ask the user to enter a value (e.g., M) for a mathematical function.

(a) Generate x values from -10 to 10 using np.linspace().


```python
import numpy as np
M = float(input("Enter a value for M: "))

# Generate x values from -10 to 10 using np.linspace()
x_values = np.linspace(-10, 10, 100)

print(f"x values from -10 to 10: \n{x_values}")
```

(b) Compute y values for:

* y= M⋅x^2

* y=M⋅sin(x)

```python
M = float(input("Enter a value for M: "))

x_values = np.linspace(-10, 10, 100)

# Compute y values for y = M * x^2
y_square = M * x_values**2

# Compute y values for y = M * sin(x)
y_sin = M * np.sin(x_values)

print(f"y values for y = M * x^2: \n{y_square}")
print(f"y values for y = M * sin(x): \n{y_sin}")
```

(c) Plot both functions in a single figure:

* Use different colors and line styles.

* Add legend, grid, and title.

```python
import matplotlib.pyplot as plt
M = float(input("Enter a value for M: "))

# Generate x values from -10 to 10 using np.linspace()
x_values = np.linspace(-10, 10, 100)

# Compute y values for y = M * x^2
y_square = M * x_values**2

# Compute y values for y = M * sin(x)
y_sin = M * np.sin(x_values)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot both functions with different colors and line styles
plt.plot(x_values, y_square, label=r'$y = M \cdot x^2$', color='blue', linestyle='-', linewidth=2)
plt.plot(x_values, y_sin, label=r'$y = M \cdot sin(x)$', color='red', linestyle='--', linewidth=2)

# Add grid, title, and labels
plt.grid(True)
plt.title('Plots of y = M * x^2 and y = M * sin(x)', fontsize=14)
plt.xlabel('x values', fontsize=12)
plt.ylabel('y values', fontsize=12)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

Q.2 Create a dataset of five subjects and scores.

(a) Convert it into a Pandas DataFrame.

```python
import pandas as pd
subjects = ['Math', 'Science', 'History', 'Literature', 'Art']
scores = [85, 92, 78, 88, 95]

# Create a Pandas DataFrame
data = {'Subject': subjects, 'Score': scores}
df = pd.DataFrame(data)

print(df)
```

(b) Plot the scores using a Seaborn bar plot with:

* Different colors for each bar.

* Annotations on top of each bar.

* Title, axis labels, and grid.

```python
import seaborn as sns
subjects = ['Math', 'Science', 'History', 'Literature', 'Art']
scores = [85, 92, 78, 88, 95]

data = {'Subject': subjects, 'Score': scores}
df = pd.DataFrame(data)

sns.set(style="whitegrid")

# Create the bar plot with Seaborn
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='viridis')

# Add annotations on top of each bar
for index, value in enumerate(df['Score']):
    bar_plot.text(index, value + 2, str(value), ha='center', va='bottom', fontsize=12)

plt.title('Scores by Subject', fontsize=16)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Score', fontsize=12)

plt.grid(True)

plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().

(a) Create a 2x2 subplot layout displaying:

* Line plot showing cumulative sum.

* Scatter plot with random noise.

```python
roll_number = 12345

np.random.seed(roll_number)

data = np.random.randn(50)

# Create a 2x2 subplot layout
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Line plot showing cumulative sum
axs[0].plot(np.cumsum(data), label='Cumulative Sum', color='blue')
axs[0].set_title('Cumulative Sum of Data')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Cumulative Sum')
axs[0].grid(True)
axs[0].legend()

# Scatter plot with random noise
noise = np.random.randn(50)  # Generate some random noise
axs[1].scatter(range(50), data + noise, color='red', label='Data with Noise')
axs[1].set_title('Scatter Plot with Random Noise')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Value')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv

Apply ‘seaborn’ library to do the following.

1. Read Total profit of all months and show it using a line plot.

2. Read all product sales data and show it using a multiline plot.

3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.lineplot(x=df["month_number"], y=df["total_profit"], marker="o", color="b", ax=axes[0])
axes[0].set_title("Total Profit per Month")
axes[0].set_xlabel("Month Number")
axes[0].set_ylabel("Total Profit")
axes[0].grid()

for column in df.columns[1:-1]:
    sns.lineplot(x=df["month_number"], y=df[column], label=column, ax=axes[1])
axes[1].set_title("Product Sales per Month")
axes[1].set_xlabel("Month Number")
axes[1].set_ylabel("Sales")
axes[1].legend()
axes[1].grid()

df.set_index("month_number").plot(kind="bar", ax=axes[2], colormap="viridis")
axes[2].set_title("Sales Data")
axes[2].set_xlabel("Month Number")
axes[2].set_ylabel("Values")
axes[2].grid(axis='y')

plt.tight_layout()
plt.show()
```
